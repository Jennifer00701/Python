```python
# 예외(exception)코드는 실행하는 중에 발생한 에러를 뜻합니다.

try : 
    x=int (input('나눌 숫자를 입력'))
    y= 10/x
    print(y)
except:
    print('예외 발생')

# 문자로 식을 나눌 수 없으니 문자열 넣으면 오류 뜸 
```

```python
y = [10,20,30]
try:
    index, x = map(int, input('나눌 숫자를 입력하세요:').split('.'))
    print(y[index]/x)
except:
    print('오류입니다')
```

```python
y=[10,20,30]
try:
    index,x = map(int, input('나눌 숫자를 입력하세요:').split('.'))
    print(y[index]/x) 
except ZeroDivisionError :
    print('숫자를 0으로 나눌 수 없음')
except IndexError:
    print("잘못된 인덱스임")
```

```python
try:
    x =int(input('나눌 숫자를 입력해라'))
    y = 10/x
    print(y)
    # raise
except:
    print("예외 발생")

else:
    print("성공")

finally:
    print("코두 실행이 종료됐음")

# x => input 입력 => y = 10/x 
# 문자열 입력 시 출력 불가하여 예외코드 발생 
# 예외코드 발생 안 할 시에 성공 else 출력
    # 출력 완료 시 finally 출력 
    #  raise 입력 시 예외코드 발생됨
```
