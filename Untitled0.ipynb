### 변수에 값을 넣는 것 => 할당

sum = 1+2 => 1+2의 계산된 결과를 sum 이라는 변수에 할당한다.  
sum => 변수명이고 "=" 은 같다는 의미가 아니라 할당(assignment) 라고 한다.  
"=" : 할당 연산자  
"+" : 산술 연산자   

+= 산술 후 바로 할당하는 문법  
/=, -=, *= 모두 가능합니다.  




**변수를 만드는 규칙**

*   영문 문자와 숫자를 사용할 수 있습니다.
*   대소문자를 구분합니다.
*   첫글자는 대문자로 쓰는걸 지향합니다.
*  문자부터 시작해야 하며 숫자부터 시작하면 안 됩니다.
*   특수문자 (+, -, *, /, $, @, % 등) 사용할 수 없습니다.
*   파이썬의 키워드 (if, for, while, and, or) 는 사용할 수 없습니다.





```python
# 변수를 사용하면서 사용 가능한 문법
# 변수명은 동일하게 선언 불가

sum = 1+2
sum = sum+3
sum = sum+4
sum = sum+5

print (sum)

sum2 = 1+2
sum2 += 3
sum2 += 4
sum2 += 5

# sum2 +=4 =sum = sum +4 랑 같다는 건데 생략된 것
```

```python
# 주의 !! 선언되지 않은 변수를 사용할 경우 에러가 발생함

# sum3 += 1 (에러발생) , sum3 라고 하면 그냥 호출만 하는 것
# sum 변수를 선언하는 것만으로는 공간을 생성 불가함

sum3 = 0 #값을 넣어줌
sum3 +=1

sum3
```

**sum3 +=1 변수 사용 시 참고  **

+=는 줄임 문법임 -> sum3 sum3+1 이 sum3 +=1 으로 왼쪽 자신의 값을 변경시킬 때 많이 사용되는 문법임 !
